## Exercise:

Write pandas code to report the number of cubic feet of volume the inventory occupies in each warehouse.

Return the result table in any order.

The query result format is in the following example.

```python
import pandas as pd

# Create the Warehouse DataFrame
data_warehouse = {
    'name': ['LCHouse1', 'LCHouse1', 'LCHouse1', 'LCHouse2', 'LCHouse2', 'LCHouse3'],
    'product_id': [1, 2, 3, 1, 2, 4],
    'units': [1, 10, 5, 2, 2, 1]
}

warehouse_df = pd.DataFrame(data_warehouse)

# Create the Products DataFrame
data_products = {
    'product_id': [1, 2, 3, 4],
    'product_name': ['LC-TV', 'LC-KeyChain', 'LC-Phone', 'LC-T-Shirt'],
    'Width': [5, 5, 2, 4],
    'Length': [50, 5, 10, 10],
    'Height': [40, 5, 10, 20]
}

products_df = pd.DataFrame(data_products)
```

```python
warehouse_df.describe()
```

```python
warehouse_df.dtypes
```

```python
warehouse_df.columns
```

```python
warehouse_df
```

```python
products_df['size'] = products_df['Width'] * products_df['Length'] * products_df['Height']
products_df
```

```python
merged_df = warehouse_df.merge(products_df, on='product_id')
merged_df
```

```python
merged_df['volume'] = merged_df['size'] * merged_df['units']
merged_df
```

```python
output_df = merged_df.groupby(['name'])['volume'].sum().reset_index(name='volume').rename(columns={"name": "warehouse_name"})
output_df
```

```python
import pandas as pd

# Create the expected DataFrame
data_expected = {
    'warehouse_name': ['LCHouse1', 'LCHouse2', 'LCHouse3'],
    'volume': [12250, 20250, 800]
}

expected_df = pd.DataFrame(data_expected)
expected_df
```

```python
pd.testing.assert_frame_equal(output_df, expected_df, check_dtype=False, check_index_type=False, check_like=True)
```

## Exercise:

Write pandas code to report the number of calls and the total call duration between each pair of distinct persons (person1, person2) where person1 < person2.

Return the result table in any order.

The query result format is in the following example.

```python
import pandas as pd
import numpy as np 

data = {
    'from_id': [1, 2, 1, 3, 3, 3, 4],
    'to_id': [2, 1, 3, 4, 4, 4, 3],
    'duration': [59, 11, 20, 100, 200, 200, 499]
}

calls_df = pd.DataFrame(data)

calls_df
```

```python
calls_df.describe()
```

```python
calls_df.dtypes
```

```python
calls_df.columns
```

```python
calls_df
```

```python
calls_df['person1'] = calls_df[['from_id', 'to_id']].min(axis=1)
calls_df
```

```python
calls_df['person2'] = calls_df[['from_id', 'to_id']].max(axis=1)
calls_df
```

```python
output_columns = ['person1', 'person2', 'duration']
results_df = calls_df[output_columns].groupby(['person1', 'person2']).agg(call_count=('duration', 'count'),
                                                             total_duration=('duration', 'sum')).reset_index()
results_df
```

```python
output_data = {
    'person1': [1, 1, 3],
    'person2': [2, 3, 4],
    'call_count': [2, 1, 4],
    'total_duration': [70, 20, 999]
}

expected_output_df = pd.DataFrame(output_data)
expected_output_df
```

```python
pd.testing.assert_frame_equal(results_df, expected_output_df, check_dtype=False, check_index_type=False, check_like=True)
```

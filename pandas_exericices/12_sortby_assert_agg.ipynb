## Exercise:

Write pandas to report the distance traveled by each user.

Return the result table ordered by travelled_distance in descending order, if two or more users traveled the same distance, order them by their name in ascending order.

The query result format is in the following example.

```python
import pandas as pd

# Users table
users_data = {'id': [1, 2, 3, 4, 7, 13, 19],
              'name': ['Alice', 'Bob', 'Alex', 'Donald', 'Lee', 'Jonathan', 'Elvis']}
users_df = pd.DataFrame(data=users_data)

# Rides table
rides_data = {'id': [1, 2, 3, 4, 5, 6, 7, 8, 9],
              'user_id': [1, 2, 3, 7, 13, 19, 7, 19, 7],
              'distance': [120, 317, 222, 100, 312, 50, 120, 400, 230]}
rides_df = pd.DataFrame(data=rides_data)
```

```python
joined_df = users_df.merge(rides_df, left_on="id", right_on="user_id", how="left")
joined_df
```

```python
joined_df["distance"] = joined_df["distance"].fillna(0).astype("int64")
```

```python
result_df = joined_df.groupby("name").agg(travelled_distance=("distance", "sum")).reset_index()
result_df
```

```python
result_df = result_df.sort_values(["travelled_distance", "name"], ascending=[False, True]).reset_index(drop=True)
result_df
```

```python
# create expected dataframe
expected_df = pd.DataFrame({
    "name": ["Elvis", "Lee", "Bob", "Jonathan", "Alex", "Alice", "Donald"],
    "travelled_distance": [450, 450, 317, 312, 222, 120, 0]
})

# compare the two dataframes
pd.testing.assert_frame_equal(result_df, expected_df, check_dtype=False)
```
